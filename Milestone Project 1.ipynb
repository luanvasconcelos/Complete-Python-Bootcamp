```python
print('Welcome to Tic Tac Toe!')
# Input taking
while True:
    player1 = str(input('Player 1: Do you want to be X or O? Please input the correct character '))
    if player1=='X' or player1=='O':
        break
# Assigning player 2 character
if player1=='X':
    player2 = 'O'
else:
    player2 = 'X'

replay = 'yes'
while replay=='yes':
    output = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    condition_x = ['X', 'X', 'X']
    condition_o = ['O', 'O', 'O']
    victory = ''
    print ('','1','|','2','|', '3',
               '\n','4','|','5','|', '6',
               '\n','7','|','8','|','9')
    print('\n')
    print ('',output[0],'|',output[1],'|', output[2],
               '\n',output[3],'|',output[4],'|', output[5],
               '\n',output[6],'|',output[7],'|', output[8])
    
    toggle = 0
    while victory=='':
        toggle += 1
        if toggle == 10:
            print('It\'s a tie!')
            replay = str(input('Do you want to play again? Type yes or no: '))
            break
            
        if toggle%2!=0:
            inp = int(input(f'Player 1 ({player1}), choose your next position: (1-9)'))
            output[inp-1] = player1
        else:
            inp = int(input(f'Player 2 ({player2}), choose your next position: (1-9)'))
            output[inp-1] = player2
        
        print ('',output[0],'|',output[1],'|', output[2],
               '\n',output[3],'|',output[4],'|', output[5],
               '\n',output[6],'|',output[7],'|', output[8])
        #HORIZONTAL CONDITIONS
        if output[0:3]==condition_o or output[0:3]==condition_x:
            victory = output[0]
            print(f'Congratulations! {victory} has won the game!')
            replay = str(input('Do you want to play again? Type yes or no: '))
            break
                         
        if output[3:6]==condition_o or output[3:6]==condition_x:
            victory = output[3]
            print(f'Congratulations! {victory} has won the game!')
            replay = str(input('Do you want to play again? Type yes or no: '))
            break
        
        if output[6:9]==condition_o or output[6:9]==condition_x:
            victory = output[6]
            print(f'Congratulations! {victory} has won the game!')
            replay = str(input('Do you want to play again? Type yes or no: '))
            break
            
        #VERTICAL CONDITIONS    
        if [output[0],output[3],output[6]]==condition_o or [output[0],output[3],output[6]]==condition_x:
            victory = output[0]
            print(f'Congratulations! {victory} has won the game!')
            replay = str(input('Do you want to play again? Type yes or no: '))
            break
        
        if [output[1],output[4],output[7]]==condition_o or [output[1],output[4],output[7]]==condition_x:
            victory = output[1]
            print(f'Congratulations! {victory} has won the game!')
            replay = str(input('Do you want to play again? Type yes or no: '))
            break
        
        if [output[2],output[5],output[8]]==condition_o or [output[2],output[5],output[8]]==condition_x:
            victory = output[2]
            print(f'Congratulations! {victory} has won the game!')
            replay = str(input('Do you want to play again? Type yes or no: '))
            break
        
        #DIAGONAL CONDITIONS
        if [output[0],output[4],output[8]]==condition_o or [output[0],output[4],output[8]]==condition_x:
            victory = output[0]
            print(f'Congratulations! {victory} has won the game!')
            replay = str(input('Do you want to play again? Type yes or no: '))
            break
        
        if [output[2],output[4],output[6]]==condition_o or [output[2],output[4],output[6]]==condition_x:
            victory = output[2]
            print(f'Congratulations! {victory} has won the game!')
            replay = str(input('Do you want to play again? Type yes or no: '))
            break
```
